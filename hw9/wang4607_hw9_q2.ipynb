```python
# commond me out if cannot display
# %matplotlib widget

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Consumer surplus and producer surplus

```python
def supply_p(q):
    return np.sqrt(q)


def supply_p_tax(q):
    return np.sqrt(q) + 2


def demand_p(q):
    return 100 * np.exp(
        -0.008 * q
    )  # <-- this is where I modified code after class it should have said q not quantities
```

## Computation
consumer surplus, the producer surplus, the revenue, and the deadweight loss in equilibrium

```python
import scipy.integrate as si
import scipy.optimize as so

before_tax_intersec_x = so.newton(lambda x: demand_p(x) - supply_p(x), x0=200)
before_tax_intersec_y = demand_p(before_tax_intersec_x)
after_tax_intersec_x = so.newton(lambda x: demand_p(x) - supply_p_tax(x), x0=200)
after_tax_intersec_y = demand_p(after_tax_intersec_x)

consumer_surplus, _ = si.quad(
    lambda x: demand_p(x) - demand_p(after_tax_intersec_x), 0, after_tax_intersec_x
)
producer_surplus, _ = si.quad(
    lambda x: supply_p(after_tax_intersec_x) - supply_p(x), 0, after_tax_intersec_x
)
revenue = after_tax_intersec_x * (
    supply_p_tax(after_tax_intersec_x) - supply_p(after_tax_intersec_x)
)
deadweight_loss, _ = si.quad(
    lambda x: demand_p(x) - supply_p(x), after_tax_intersec_x, before_tax_intersec_x
)

print(f"Consumer surplus: {consumer_surplus}")
print(f"Producer surplus: {producer_surplus}")
print(f"Revenue: {revenue}")
print(f"Deadweight loss: {deadweight_loss}")
```

## Plot

```python
deadweight_line = [
    (after_tax_intersec_x, after_tax_intersec_x),
    (supply_p(after_tax_intersec_x), after_tax_intersec_y),
]
revenue_upper = [
    (0, after_tax_intersec_x),
    [demand_p(after_tax_intersec_x), demand_p(after_tax_intersec_x)],
]
revenue_lower = [
    (0, after_tax_intersec_x),
    [supply_p(after_tax_intersec_x), supply_p(after_tax_intersec_x)],
]
zero_line = [(0, 0), [supply_p(0), demand_p(0)]]

quantities = np.linspace(0, 250, 101)
supply = supply_p(quantities)
supply_tax = supply_p_tax(quantities)
demand = demand_p(quantities)

fig = plt.figure(figsize=[20, 15])
plt.plot(quantities, demand, label="demand function")
plt.plot(quantities, supply, label="supply function before tax")
plt.plot(quantities, supply_tax, label="supply function after tax")
plt.plot(*deadweight_line, "k:")
plt.plot(*revenue_upper, "k:")
plt.plot(*revenue_lower, "k:")
plt.plot(*zero_line, "k:")

_x = np.linspace(0, after_tax_intersec_x, 1000)
_y1 = demand_p(_x)
_y2 = np.full_like(_y1, after_tax_intersec_y)
plt.fill_between(_x, _y1, _y2, alpha=0.5)
plt.text(np.mean(_x) - 50, np.mean((_y1 + _y2) / 2), "Consumer surplus")

_x = np.linspace(0, after_tax_intersec_x, 1000)
_y1 = np.full_like(_x, after_tax_intersec_y)
_y2 = np.full_like(_x, supply_p(after_tax_intersec_x))
plt.fill_between(_x, _y1, _y2, alpha=0.5)
plt.text(np.mean(_x), np.mean((_y1 + _y2) / 2), "Revenue")

_x = np.linspace(after_tax_intersec_x, before_tax_intersec_x, 100)
_y1 = np.full_like(_x, demand_p(_x))
_y2 = np.full_like(_x, supply_p(_x))
plt.fill_between(_x, _y1, _y2, alpha=0.5)
plt.annotate(
    "Deadweight loss", 
    (np.mean(_x), np.mean((_y1 + _y2) / 2)),
    xytext=(np.mean(_x) + 5, np.mean((_y1 + _y2) / 2) + 5),
    arrowprops = {"arrowstyle": "->"}
)

_x = np.linspace(0, after_tax_intersec_x, 1000)
_y1 = np.full_like(_x, before_tax_intersec_y)
_y2 = supply_p(_x)
plt.fill_between(_x, _y1, _y2, alpha=0.5)
plt.text(np.mean(_x) - 50, np.mean((_y1 + _y2) / 2), "Producer surplus")

plt.legend()
plt.xlabel("Q")
plt.ylabel("P")
plt.show()
```
