The purpose of this project is to create a simulation of how COVID-19 can spread through a community given certain parameters.

1. Investigate the effect of the length of asymptomatic window. Specifically, plot the cumulative number of infections over time for the case that L=3 and compare it to the case when L=6.

```python
simulator_L3 = Covid19Simulator(L=3)
for s in range(10):
    simulator_L3.reset(s)
    for _ in range(50):
        simulator_L3.simulate_one_day()
        
simulator_L6 = Covid19Simulator(L=6)
for s in range(10):
    simulator_L6.reset(s)
    for _ in range(50):
        simulator_L6.simulate_one_day()

log_L3 = simulator_L3.get_log()
log_L3.insert(0, "L", 3)
log_L6 = simulator_L6.get_log()
log_L6.insert(0, "L", 6)
log = pd.concat([log_L3, log_L6])

sns.lineplot(x="date", y="cumulative infected number", hue="L", data=log)
```

Explanation: When “asymptomatic window” becomes longer, the cumulative infected number becomes larger.

5. (Unique Modification) If the chance of infection becomes higher, specifically for `I=0.35` and `I=0.7`, the cumulative infected number, cumulative test number, daily test number, daily increased infected number, and daily test positive rate will change. 

```python
simulator1 = Covid19Simulator(I=0.35)
for s in range(10):
    simulator1.reset(s)
    for _ in range(50):
        simulator1.simulate_one_day()
        
simulator2 = Covid19Simulator(I=0.7)
for s in range(10):
    simulator2.reset(s)
    for _ in range(50):
        simulator2.simulate_one_day()

log1 = simulator1.get_log()
log1.insert(0, "I", 0.35)
log2 = simulator2.get_log()
log2.insert(0, "I", 0.7)
log = pd.concat([log1, log2])

sns.lineplot(x="date", y="cumulative infected number", hue="I", data=log)
```

Explanation: When the infection rate becomes higher, the cumulative infected number becomes larger.

```python
sns.lineplot(x="date", y="infected number today", hue="I", data=log)
```

Explanation: When the infection rate becomes higher, the maximum of today's infected number is larger.

```python
sns.lineplot(x="date", y="tested number today", hue="I", data=log)
```

Explanation: When the infection rate becomes higher, the today's tested number changes more dramatically. It demonstrates a periodical change.

```python
sns.lineplot(x="date", y="tested positive rate today", hue="I", data=log)
```

Explanation: When the infection rate becomes higher, the today's tested positive rate becomes higher.

```python
sns.lineplot(x="date", y="cumulative tested number", hue="I", data=log)
```

Explanation: When the infection rate becomes higher, the cumulative tested number becomes smaller in long term. This is kind of counter intuition. I guess this happens because people will isolate after they test positive. As a result, they won't go to test during quarantine. This is also shown in the graph of "tested number today".
